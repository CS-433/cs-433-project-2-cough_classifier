```python
# box plots
print(X.columns)
for top, features in FEATURES.items():
    print(top, features)
    if "SNR" in features and not "SNR" in X.columns:
        features.remove("SNR")
    X[features].plot(kind="box", figsize=(15, 15), logy="sym")
```

```python
N = len(X.columns)

fig, axs = plt.subplots(10, 7, figsize=(50, 40))

for i, feature in enumerate(X.columns):
    subplot = axs[i // 7, i % 7]
    X[feature].plot(kind="hist", bins=200, logy="sym", ax=subplot)
```
